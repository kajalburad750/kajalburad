```python
import matplotlib.pyplot as plt
```

```python
y_salary = [20,30,40,50]
x_cgpa = [2,3,4,5]
```

```python
plt.plot(x_cgpa,y_salary,"o-g")
plt.xlabel("cgpa")
plt.ylabel("salary")
plt.title("CGPA and Salary Distribution")
plt.show()
```
